```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def flip_binary_tree_clockwise(root):
    if root is None:
        return None

    if root.left is None and root.right is None:
        return root

    flipped_left = flip_binary_tree_clockwise(root.left)
    flipped_right = flip_binary_tree_clockwise(root.right)

    root.left = flipped_right
    root.right = flipped_left

    return root

def display_binary_tree(root):
    if root is None:
        return

    print(root.val, end=" ")
    display_binary_tree(root.left)
    display_binary_tree(root.right)


root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

print("Original binary tree:")
display_binary_tree(root)
flipped_root = flip_binary_tree_clockwise(root)

print("\nFlipped binary tree:")
display_binary_tree(flipped_root)
```

Given a binary tree, print all its root-to-leaf paths without using recursion. For example, consider the following Binary Tree.

Input:

        6
     /    \
    3      5
  /   \     \
 2     5     4
     /   \
    7     4

Output:

There are 4 leaves, hence 4 root to leaf paths -
  6->3->2
  6->3->5->7
  6->3->5->4
  6->5>4

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def root_to_leaf_path(root):
    if root is None:
        return
    stack = []
    stack.append((root, str(root.val)))

    while stack:
        node, path = stack.pop()

        if node.left is None and node.right is None:
            print(path)
        
        if node.right is not None:
            stack.append((node.right, path + "->" + str(node.right.val)))
        
        if node.left is not None:
            stack.append((node.left, path + "->" + str(node.left.val)))


root = TreeNode(6)
root.left = TreeNode(3)
root.right = TreeNode(5)
root.left.left = TreeNode(2)
root.left.right = TreeNode(5)
root.right.right = TreeNode(4)
root.left.right.left = TreeNode(7)
root.left.right.right = TreeNode(4)

root_to_leaf_path(root)
```

GivenÂ Preorder, Inorder and PostorderÂ traversals of some tree. Write a program to check if they all are of the same tree.

**Examples:**

Input : 

        Inorder -> 4 2 5 1 3
        Preorder -> 1 2 4 5 3
        Postorder -> 4 5 2 3 1
Output : 

Yes
Explanation : 

All of the above three traversals are of
the same tree 

                           1
                         /   \
                        2     3
                      /   \
                     4     5

Input : 

        Inorder -> 4 2 5 1 3
        Preorder -> 1 5 4 2 3
        Postorder -> 4 1 2 3 5
Output : 

No

</aside>

```python
def check_for_same_tree(preorder, inorder, postorder):
    if len(preorder) != len(inorder) or len(preorder) != len(postorder):
        return False

    def check_for_same_tree_recursive(preorder, inorder, postorder):
        if not preorder:
            return True

        if preorder[0] != postorder[-1]:
            return False

        idx = inorder.index(preorder[0])

        left_preorder = preorder[1:idx+1]
        left_inorder = inorder[:idx]
        left_postorder = postorder[:idx]

        right_preorder = preorder[idx+1:]
        right_inorder = inorder[idx+1:]
        right_postorder = postorder[idx:-1]

        return check_for_same_tree_recursive(left_preorder, left_inorder, left_postorder) and \
               check_for_same_tree_recursive(right_preorder, right_inorder, right_postorder)

    return check_for_same_tree_recursive(preorder, inorder, postorder)

# Example usage
inorder = [4, 2, 5, 1, 3]
preorder = [1, 2, 4, 5, 3]
postorder = [4, 5, 2, 3, 1]
print(check_for_same_tree(preorder, inorder, postorder)) 

inorder = [4, 2, 5, 1, 3]
preorder = [1, 5, 4, 2, 3]
postorder = [4, 1, 2, 3, 5]
print(check_for_same_tree(preorder, inorder, postorder))  
```
