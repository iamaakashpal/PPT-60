Question-1:

Given preorder of a binary tree, calculate its **[depth(or height)](https://www.geeksforgeeks.org/write-a-c-program-to-find-the-maximum-depth-or-height-of-a-tree/)** [starting from depth 0]. The preorder is given as a string with two possible characters.

1. ‘l’ denotes the leaf
2. ‘n’ denotes internal node

The given tree can be seen as a full binary tree where every node has 0 or two children. The two children of a node can ‘n’ or ‘l’ or mix of both.

**Examples :**

    Input  : nlnll
    Output : 2
    Explanation :

![img](https://pwskills.notion.site/image/https%3A%2F%2Fmedia.geeksforgeeks.org%2Fwp-content%2Fuploads%2Fbtree1.png?id=b9cd6bb1-1a24-45cf-a2ec-d9a665b7fd24&table=block&spaceId=6fae2e0f-dedc-48e9-bc59-af2654c78209&width=580&userId=&cache=v2)

    Input  : nlnnlll
    Output : 3

![img](https://pwskills.notion.site/image/https%3A%2F%2Fmedia.geeksforgeeks.org%2Fwp-content%2Fuploads%2Fdia2.png?id=604e28b1-a250-4e4f-9416-91e5c4545e60&table=block&spaceId=6fae2e0f-dedc-48e9-bc59-af2654c78209&width=580&userId=&cache=v2)
</aside>

```python
def calculate_height(order):
    stack = []
    depth = -1

    for char in order:
        if char == 'n':
            stack.append(char)
        elif char == 'l':
            if not stack:
                continue
            while stack and stack[-1] == 'l':
                stack.pop()
            if stack:
                stack.pop() 
            stack.append('l')

        depth = max(depth, len(stack) - 1)

    return depth

calculate_height("nlnll")
```

Question-2:

Given a Binary tree, the task is to print the **left view** of the Binary Tree. The left view of a Binary Tree is a set of leftmost nodes for every level.

**Examples:**

***Input:***

            4

          /   \

        5     2

             /   \

            3     1

           /  \

          6    7

***Output:** 4 5 3 6*

**Explanation:**

![img](https://pwskills.notion.site/image/https%3A%2F%2Fmedia.geeksforgeeks.org%2Fwp-content%2Fcdn-uploads%2Fleft-view.png?id=f7c11345-294e-45c8-99b2-eb1553a0c93d&table=block&spaceId=6fae2e0f-dedc-48e9-bc59-af2654c78209&width=380&userId=&cache=v2)

***Input:***

                    1

                  /   \

                2       3

                 \

                   4

                     \

                        5

                           \

                             6

**Output:** 1 2 4 5 6

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def left_view(root):
    if root is None:
        return

    queue = [(root, 0)] 
    map = {} 

    while queue:
        node, level = queue.pop(0)

        if level not in map:
            map[level] = node.val

        if node.left:
            queue.append((node.left, level + 1))

        if node.right:
            queue.append((node.right, level + 1))

    for level in sorted(map.keys()):
        print(map[level], end=" ")

root = TreeNode(4)
root.left = TreeNode(5)
root.right = TreeNode(2)
root.right.left = TreeNode(3)
root.right.right = TreeNode(1)
root.right.left.left = TreeNode(6)
root.right.left.right = TreeNode(7)

left_view(root)
```

Question-3:

Given a Binary Tree, print the Right view of it.

The right view of a Binary Tree is a set of nodes visible when the tree is visited from the Right side.

**Examples:**

**Input:**

         1

      /     \

   2         3

/   \       /  \

4     5   6    7

             \

               8

**Output**: 

Right view of the tree is 1 3 7 8

**Input:**

         1

       /

    8

  /

7

**Output**: 

Right view of the tree is 1 8 7

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def right_view(root):
    if root is None:
        return

    queue = [(root, 0)]
    map = {}

    while queue:
        node, level = queue.pop(0)

        map[level] = node.val

        if node.left:
            queue.append((node.left, level + 1))

        if node.right:
            queue.append((node.right, level + 1))

    for level in sorted(map.keys()):
        print(map[level], end=" ")

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)
root.right.left.right = TreeNode(8)

right_view(root)
```

Question-4:

Given a Binary Tree, The task is to print the **bottom view** from left to right. A node **x** is there in output if x is the bottommost node at its horizontal distance. The horizontal distance of the left child of a node x is equal to a horizontal distance of x minus 1, and that of a right child is the horizontal distance of x plus 1.

**Examples:**

**Input:**

             20

           /     \

        8         22

    /      \         \

5         3        25

        /    \

   10       14

**Output:** 5, 10, 3, 14, 25.

**Input:**

             20

           /     \

        8         22

    /      \      /   \

 5         3    4     25

         /    \

     10       14

**Output:**

5 10 4 14 25.

**Explanation:**

If there are multiple bottom-most nodes for a horizontal distance from the root, then print the later one in the level traversal.

**3 and 4** are both the bottom-most nodes at a horizontal distance of 0, we need to print 4.

</aside>

```python
from collections import deque

class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def bottom_view(root):
    if root is None:
        return

    map = {}

    queue = deque([(root, 0)])

    while queue:
        node, hd = queue.popleft()

        map[hd] = node.val

        if node.left:
            queue.append((node.left, hd - 1))

        if node.right:
            queue.append((node.right, hd + 1))

    for hd in sorted(map.keys()):
        print(map[hd], end=" ")

root = TreeNode(20)
root.left = TreeNode(8)
root.right = TreeNode(22)
root.left.left = TreeNode(5)
root.left.right = TreeNode(3)
root.right.right = TreeNode(25)
root.left.right.left = TreeNode(10)
root.left.right.right = TreeNode(14)

bottom_view(root)
```
